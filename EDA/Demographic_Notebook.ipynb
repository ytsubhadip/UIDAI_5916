```python
## remove space and symbol
import re
import pandas as pd

def clean_name(x):
    # 1. Check if the value is missing (NaN)
    if pd.isna(x):
        return x

    # 2. Convert to string and make it lowercase
    x = str(x).lower()

    # 3. Remove symbols: replace anything that is NOT a letter (a-z)
    # or a space (\s) with a single space
    x = re.sub(r'[^a-z\s]', ' ', x)  # removes (), ., &, etc.

    # 4. Remove extra spaces: replace multiple spaces with one
    # and remove leading/trailing spaces
    x = re.sub(r'\s+', ' ', x).strip()

    return x
```

```python
# Yeh dictionary "Galat/Duplicate Naam" ko "Sahi Naam" se map karegi
bihar_district_mapping = {
 # --- Champaran (Hindi/Variations -> English) ---
    'Purbi Champaran': 'East Champaran',
    'Purba Champaran': 'East Champaran',
    'Pashchim Champaran': 'West Champaran',

    # --- Aurangabad (Removing 'BH' codes) ---
    'Aurangabad(bh)': 'Aurangabad',
    'Aurangabad(BH)': 'Aurangabad',

    # --- Spelling Corrections (Typos) ---
    'Samstipur': 'Samastipur',   # Missing 'a'
    'Sheikpura': 'Sheikhpura',   # Missing 'h'
    'Purnea': 'Purnia',          # Alternative spelling

    # --- Old Names ---
    'Monghyr': 'Munger',         # British era name

    # --- Kaimur / Bhabua Standardization ---
    'Bhabua': 'Kaimur (Bhabua)', # Maps HQ name to District name

    # --- Capitalization Fixes ---
    'kishanganj': 'Kishanganj'
}
```
